```python
import pandas as pd

train = pd.read_csv("train.csv")
meal = pd.read_csv("meal_info.csv")
center = pd.read_csv("fulfilment_center_info.csv")

print("All files loaded successfully ✅")
```

```python
data = pd.merge(train, meal, on="meal_id")
```

```python
data = pd.merge(data, center, on="center_id")
```

```python
data.head()
```

```python
data = data.drop(["emailer_for_promotion", "homepage_featured"], axis=1)
```

```python
data.head()
```

```python
data_sorted = data.sort_values(by="num_orders", ascending=False)
```

```python
data_top10k = data_sorted.head(10000)
```

```python
data_top10k.shape
```

```python
data.info()
data.isnull().sum()
```

```python
data_top10k = data_top10k.copy()
```

```python
# Discount
data_top10k["discount_percent"] = (
    (data_top10k["base_price"] - data_top10k["checkout_price"]) 
    / data_top10k["base_price"]
) * 100

# Revenue
data_top10k["revenue"] = data_top10k["checkout_price"] * data_top10k["num_orders"]

# Demand Level
data_top10k["demand_level"] = pd.cut(
    data_top10k["num_orders"],
    bins=[0, 5000, 15000, 30000],
    labels=["Low", "Medium", "High"]
)
```

```python
data_top10k.head()
```

```python
data_top10k = data_top10k.reset_index(drop=True)
```

```python
data_top10k.head()
```

```python
data_top10k.to_csv("cleaned_data.csv", index=False)
```

```python
avg_revenue = data_top10k['revenue'].mean()

data_top10k['revenue_category'] = data_top10k['revenue'].apply(
    lambda x: 'High' if x > avg_revenue else 'Low'
)
```

```python
data_top10k['week_type'] = data_top10k['week'].apply(
    lambda x: 'Start' if x <= 17 else ('Mid' if x <= 35 else 'End')
)
```

```python
avg_orders = data_top10k['num_orders'].mean()

data['meal_popularity'] = data_top10k['num_orders'].apply(
    lambda x: 'Popular' if x > avg_orders else 'Not Popular'
)
```

```python
data_top10k.head()
```

```python
data_top10k['profit'] = (data_top10k['checkout_price'] - data_top10k['base_price']) * data_top10k['num_orders']
```

```python
data_top10k.head()
```

```python
data_top10k.rename(columns={'id': 'order_id'}, inplace=True)
```
